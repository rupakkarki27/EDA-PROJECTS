# Plotly and Cufflinks

```python
import pandas as pd
import numpy as np
import cufflinks as cf
%matplotlib inline
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Categories':['A', 'B', 'C'],
                    'Values':[30, 40, 50]})
```

```python
df2
```

```python
df.iplot(title='Interactive')
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
df2.iplot(kind='bar', x='Categories', y='Values')
```

```python
df.mean().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],
                   'y':[5,10,15,20,25],
                   'z':[100,200,300,400,500]})
```

```python
df3.iplot(kind='surface')
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble',x='A', y='B', size='C',colorscale='rdylbu')
```
